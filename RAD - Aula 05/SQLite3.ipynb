```python
# SQLite já vem com Python (não precisa instalar)
import sqlite3

print("SQLite carregado com sucesso!")
```

### 2. Criando conexão com banco

```python
# Criando (ou conectando) ao banco
conn = sqlite3.connect("empresa.db")

# Cursor (permite executar comandos SQL)
cursor = conn.cursor()

print("Banco conectado!")
```

### 3. Criando tabelas

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS funcionarios (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nome TEXT NOT NULL,
    idade INTEGER CHECK(idade > 0 AND idade < 120),
    cidade TEXT DEFAULT 'Não informada',
    salario REAL CHECK(salario >= 0),
    data_contratacao TEXT
)
""")

conn.commit()

print("Tabela criada!")
```

### 4. Inserindo dados (CREATE / INSERT)

```python
cursor.execute("""
INSERT INTO funcionarios (nome, idade, cidade, salario, data_contratacao)
VALUES (?, ?, ?, ?, ?)
""", ("Ana", 28, "Recife", 3500.00, "2024-01-10"))

conn.commit()

print("Dados inseridos!")
```

### 5. Inserindo múltiplos dados

```python
dados = [
    ("João", 35, "Palmares", 4200, "2023-05-20"),
    ("Maria", 30, "Caruaru", 3900, "2022-03-15"),
    ("Carlos", 40, "Recife", 5000, "2021-07-01")
]

cursor.executemany("""
INSERT INTO funcionarios (nome, idade, cidade, salario, data_contratacao)
VALUES (?, ?, ?, ?, ?)
""", dados)

conn.commit()

print("Vários dados inseridos!")
```

### 6. Consultando dados (SELECT)

```python
cursor.execute("SELECT * FROM funcionarios")

resultados = cursor.fetchall()

for linha in resultados:
    print(linha)
```

### 7. Filtros (WHERE)

```python
cursor.execute("""
SELECT nome, salario
FROM funcionarios
WHERE salario > 4000
""")

for linha in cursor.fetchall():
    print(linha)
```

### 8. Atualizando dados (UPDATE)

```python
cursor.execute("""
UPDATE funcionarios
SET salario = salario * 1.1
WHERE cidade = 'Recife'
""")

conn.commit()

print("Salários atualizados!")
```

### 9. Deletando dados (DELETE)

```python
cursor.execute("""
DELETE FROM funcionarios
WHERE nome = 'Carlos'
""")

conn.commit()

print("Registro removido!")
```

### 10. Usando funções SQL (GROUP BY)

```python
cursor.execute("""
SELECT cidade, AVG(salario)
FROM funcionarios
GROUP BY cidade
""")

for linha in cursor.fetchall():
    print(linha)
```

### 11. Integração com Pandas

```python
import pandas as pd

df = pd.read_sql_query("SELECT * FROM funcionarios", conn)

df
```

### 12. Salvando DataFrame no banco

```python
df.to_sql("funcionarios_backup", conn, if_exists="replace", index=False)

print("Tabela criada via Pandas!")
```

### 13. Criando segunda tabela (relacionamento)

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS departamentos (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nome TEXT NOT NULL
)
""")

cursor.execute("""
ALTER TABLE funcionarios ADD COLUMN departamento_id INTEGER
""")

conn.commit()
```

### 14. JOIN (relacionamento entre tabelas)

```python
cursor.execute("""
SELECT f.nome, f.salario, d.nome
FROM funcionarios f
LEFT JOIN departamentos d
ON f.departamento_id = d.id
""")

for linha in cursor.fetchall():
    print(linha)
```

### 15. Boas práticas (IMPORTANTÍSSIMO)

```python
# Sempre use parâmetros (evita SQL Injection)

nome = "Ana"

cursor.execute("SELECT * FROM funcionarios WHERE nome = ?", (nome,))
print(cursor.fetchall())
```

```python
conn.close()
print("Conexão encerrada!")
```
